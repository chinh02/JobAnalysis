```python
def get_url_1():
    urls = []
    for x in range (1,53):
        driver.implicitly_wait(5)
        driver.get("https://www.topcv.vn/viec-lam-it?page={}".format(x))
        main_page_source = driver.page_source
        soup = BeautifulSoup(main_page_source,'html.parser')
        boxes = soup.find_all("h3",{"class":"title"})
        for box in boxes:
            try:
                url = box.find("a",{"target":"_blank"},href = True)
                urls.append(url['href'])
            except TypeError: continue
    return urls
```

```python
def extract_record(url_detail):
    driver.get(url_detail)
    driver.implicitly_wait(5)
    detail_pagesource = driver.page_source
    detail_soup = BeautifulSoup(detail_pagesource,'html.parser')
    try:
        boxInfo = detail_soup.find('div',{'class':'box-main'}).find_all('span')
        salary = boxInfo[0].text.strip()
        amount = boxInfo[1].text.strip()
        jobType = boxInfo[2].text.strip()
        position = boxInfo[3].text.strip()
        gender = boxInfo[4].text.strip()
        experience = boxInfo[5].text.strip()
    except AttributeError:
        boxInfo = ' '
        salary = ' '
        amount = ' '
        jobType = ' '
        position = ' '
        gender = ' '
        experience = ' '  
    try: 
        jobTitle = detail_soup.find('h1',{'class':'job-title text-highlight bold'}).text.strip()
    except AttributeError:
        try:
            jobTitle = detail_soup.find('h2',{'class':'title'}).text.strip()
        except AttributeError:
            jobTitle = ' '
    try:
        companyName = detail_soup.find('div',{'class':'company-title'}).text.strip()
        companyLink = detail_soup.find('div',{'class':'company-title'}).find('a', href = True)
        companyLink = companyLink['href']
    except:
        try:
            companyName = detail_soup.find('div',{'id':'company-name'}).text.strip()
            companyLink = detail_soup.find('div',{'id':'company-name'}).find('a', href = True)
            companyLink = companyLink['href']
        except:
            companyName =' '
            companyLink = ' '
    try:
        address = detail_soup.find('div',{'style': 'margin-bottom: 10px'}).text.strip()
        address = address.replace("-","")
        cleaned_address = re.findall("^.*?(?=:)",address)
        cleaned_address = ' '.join(cleaned_address).strip()
    except: 
        address = ' '
        cleaned_address = ' '
    try:
        desc =  detail_soup.find('div',{'class':'job-data'}).text.strip()
    except AttributeError:
        desc = ''
    try:
        driver.get(companyLink)
        company_pagesource = driver.page_source
        company_soup = BeautifulSoup(company_pagesource,'html.parser')
    
        companySize = company_soup.find('p',{'class':'company-size'}).text.strip()
    except:
        companySize = ' '
    details=(jobTitle, companyName,salary, amount,jobType,position,gender,experience,cleaned_address,companySize,desc)
    return details
```

```python
driver = uc.Chrome()
records = []
for url in urls:
    records.append(extract_record(url))
```

```python
with open ('topcv103.csv','w',newline='',encoding='UTF-16') as f:
    writer = csv.writer(f)
    writer.writerow(['jobTitle','companyName','salary','amount','jobType','position','gender','experience','address','companySize','desc'])
    writer.writerows(map(lambda x: [x], records))
```
