```python
def get_url():
    url = []
    for x in range(1,62):
        driver.get("https://itviec.com/it-jobs?page={}".format(x))
        driver.implicitly_wait(10)
        pagesource = driver.page_source
        soup = BeautifulSoup(pagesource,'html.parser')
        super_hot_jobs = soup.find_all('div',{'class':'job super-hot-job'})
        jobs = soup.find_all('div',{'class':'job'})
        for job in jobs:
            urls.append(job['data-search--job-selection-job-slug-value'])
        for hot_job in super_hot_jobs:
            urls.append(hot_job['data-search--job-selection-job-slug-value'])
    return urls
```

```python
def extract_record():
    time.sleep(1)
    detail_pagesource = driver.page_source
    detail_soup = BeautifulSoup(detail_pagesource,'html.parser')
    jobTitle = detail_soup.find("h1",{"class":"job-details__title"}).text.strip()
    companyName = detail_soup.find("div",{"class":"job-details__sub-title"}).text.strip()
    try:
        address = detail_soup.find("div",{"class":"svg-icon"}).text.strip()
        list_address = list(address.split(','))
        cleaned_address = str(list_address[-1])
    except AttributeError:
        list_address=''
        cleaned_address=''
    jobType = detail_soup.find('div',{'class':"svg-icon__box d-flex align-items-center"}).text.strip()
    try:
        salary = detail_soup.find("span",{"class":"svg-icon svg-icon--rounded svg-icon--green"}).text.strip()
    except:
        salary=''
    details=(jobTitle, companyName,salary,jobType,cleaned_address)
    return details
```

```python
def auto_click():
    wait = WebDriverWait(driver, 10)
    records = []
    for x in range(1,3):
        driver.get("https://itviec.com/it-jobs?page={}".format(x))
        driver.implicitly_wait(5)
        jobs = driver.find_elements(By.XPATH,"//div[@class='job']")
        hot_jobs = driver.find_elements(By.XPATH,"//div[@class='job super-hot-job']")
        time.sleep(3)
        for job in jobs:
            try:
                driver.execute_script("arguments[0].scrollIntoView();", job)
                job.click()
            except:
                wait.until(EC.staleness_of(job))
                driver.execute_script("arguments[0].scrollIntoView();", job)
                job.click()
            driver.refresh()
            driver.implicitly_wait(2)
            records.append(extract_record())
        for hot_job in hot_jobs:
            try:
                driver.execute_script("arguments[0].scrollIntoView();", hot_job)
                job.click()
            except:
                wait.until(EC.staleness_of(job))
                driver.execute_script("arguments[0].scrollIntoView();", hot_job)
                job.click()
            driver.refresh()
            driver.implicitly_wait(2)
            records.append(extract_record())
    return records
```
